```python
import pandas as pd
import numpy as np
```

```python
electoralVotes = pd.read_csv('data/2016ElectoralVotesLocal.csv')
stateVotes = pd.read_csv('data/state-votes.csv')
```

```python
stateVotes = pd.read_csv('data/state-votes.csv')
```

```python
statePopulations = pd.read_csv('data/StatePopulations.csv')
```

```python
df = pd.merge(statePopulations, stateVotes[["Trump", "Clinton"]], left_index=True, right_index=True)
```

```python
df = pd.merge(df, electoralVotes[["Votes", "Winning Party"]], left_index=True, right_index=True)
```

```python
df
```

```python
df.to_csv(r'data/state-winners-populations.csv')
```

```python
df["Winning Votes"] = np.array(list(map(lambda x, y: max(x, y),
                                        df["Trump"], df["Clinton"])))
```

```python
df["Total Votes"] = np.array(list(map(lambda x, y: x + y,
                                     df["Trump"], df["Clinton"])))
```

```python
ne = sum(df["Votes"])
ne
```

```python
nationalPopulation = sum(df["2018 Population"])
nationalPopulation
```

```python
totalVotesCast = sum(df["Total Votes"])
totalVotesCast
```

```python
def calculate_weight(nei, pi, n):
    return (nei / ne) * (n / pi)
```

```python
state_abbrev = pd.read_csv('data/state-abbreviations.csv')
```

```python
df = pd.merge(df, state_abbrev[["Abbreviation"]], left_index=True, right_index=True)
```

```python
df
```

```python
df_weights = df
```

```python
df_weights["Potenial Ballot Weight"] = np.array(list(map(lambda pi, nei: calculate_weight(nei, pi, nationalPopulation),
                                     df_weights["2018 Population"], df_weights["Votes"])))
```

```python
df_weights["Cast Ballot Weight"] = np.array(list(map(lambda pi, nei: calculate_weight(nei, pi, totalVotesCast),
                                     df_weights["Total Votes"], df_weights["Votes"])))
```

```python
df_weights["Final Ballot Weight"] = np.array(list(map(lambda pi, nei: calculate_weight(nei, pi, totalVotesCast),
                                     df_weights["Winning Votes"], df_weights["Votes"])))
```

```python
df_weights
```

```python
df_weights.to_csv(r'data/state-wieghts.csv')
```
