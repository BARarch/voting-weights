```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'data/state-weights.csv')
```

```python
abbrevs = df["Abbreviation"]
states = df["State"]
weights = df["Cast Ballot Weight"]
```

# Cast Ballot Weight Comparison Table
A vote cast in Alaska has 1.61 times more power than one cast in California before the resultes of the election are know

```python
voteWeights = [{ab: i / j for ab, j in zip(abbrevs, weights)} for state, i in zip(states, weights)]
table = pd.DataFrame(voteWeights)
table = pd.merge(df[["State"]], table, left_index=True, right_index=True)
table
```

```python
abbrevs = df["Abbreviation"]
states = df["State"]
weights = df["Final Ballot Weight"]
```

# Final Ballot Weight Comparison Table
A vote cast in California for the winning candidate had 1.45 times more power than one cast for the winning candidate in Alabama. We know this after the results of the election are known

```python
voteWeights = [{ab: i / j for ab, j in zip(abbrevs, weights)} for state, i in zip(states, weights)]
table = pd.DataFrame(voteWeights)
table = pd.merge(df[["State"]], table, left_index=True, right_index=True)
table
```
