```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'data/state-weights.csv')
```

```python
abbrevs = df["Abbreviation"]
states = df["State"]
weights = df["Cast Ballot Weight"]
```

# Cast Ballot Weight Comparison Table
A vote cast in Alaska has 1.61 times more power than one cast in California before the results are known

```python
voteWeightsOrdered = [list(i / j for j in weights) for i in weights]
table = pd.DataFrame(voteWeightsOrdered, columns=list(abbrevs), index=list(states))
table
```

```python
abbrevs = df["Abbreviation"]
states = df["State"]
weights = df["Final Ballot Weight"]
```

# Final Ballot Weight Comparison Table
A vote cast in California for the winning candidate had 1.46 times more power than one cast for the winning candidate in Alabama. We know this after the results of the election are known

```python
voteWeightsOrdered = [list(i / j for j in weights) for i in weights]
table = pd.DataFrame(voteWeightsOrdered, columns=list(abbrevs), index=list(states))
table
```
